# Задание №1

```python
# График функции в диапазоне от -5 до 5 с шагом 1
x = np.arange(-5, 5, 0.1) # диапазон
y = -1 * np.exp(np.sin(x)) + 2 * x * np.sin(np.exp(x - 5)) + 4 * np.sin(x - 1) # функция
pylab.plot(x, y) 
pylab.xlabel('x')
pylab.ylabel('y(x)')
```

# Задание №2

```python
# Найдите два различных нуля функции с точностью до 0,001
def f(x):
    return -1 * np.exp(np.sin(x)) + 2 * x * np.sin(np.exp(x - 5)) + 4 * np.sin(x-1)
```

```python
xlist = np.arange(-5, 5, 0.1)

plt.figure(num=1, dpi=120)
plt.plot(xlist, f(xlist), label="function")
plt.plot(xlist, xlist*0, "--", label="y=0")
plt.legend()
```

```python
root = fsolve(f, [-2, 2])
root
```

```python
# Найдите два различных минимума функции вашего варианта с точностью до 0,001
def func(x):
    return -1 * np.exp(np.sin(x)) + 2 * x * np.sin(np.exp(x - 5)) + 4 * np.sin(x-1)

xlist = np.linspace(-8, 5, 100)

plt.figure(num=1, dpi=120)
plt.plot(xlist, f(xlist), label="function")
```

```python
minimum_1 = fmin(func, 2)
minimum_2 = fmin(func, -6)
```

```python
minimum_1, minimum_2 # точки минимумов
```
